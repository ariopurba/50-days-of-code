```python
from pathlib import Path
import shutil

# Path ke dataset asli dan folder tujuan
src_path = Path('./dataset/Uang Baru')
dst_path = Path('./sampel')
dst_path.mkdir(exist_ok=True)

# Ambil 2 sampel dari setiap kelas/folder
for class_folder in src_path.iterdir():
    if class_folder.is_dir():
        dst_class = dst_path / class_folder.name
        dst_class.mkdir(exist_ok=True)
        images = sorted([img for img in class_folder.iterdir() if img.is_file()])[:10]
        for img in images:
            shutil.copy(img, dst_class / img.name)

print("Sampel berhasil dibuat di folder 'sampel'.")
```
